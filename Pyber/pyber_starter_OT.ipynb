# Pyber Analysis

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_df, city_df, how='left', on='city')

# Display the data table for preview
pyber_df.head(10)
```

```python
#  Output mergered data in the csv-file
pyber_df.to_csv("data/GroupCityType.csv", index=False, header=True)
```

## Bubble Plot of Ride Sharing Data

```python
# Build a Bubble Plot that showcases the relationship between four key variables:
# Average Fare ($) Per City - y_axis
# Total Number of Rides Per City - x_axis
# Total Number of Drivers Per City - size
# City Type (Urban, Suburban, Rural) - color(legend)

# City Type 
city_types = list(city_df["type"].unique())
# ['Urban', 'Suburban', 'Rural']

# Define colors in the Pyber color scheme (Gold, SkyBlue, and Coral)
colors = ["Coral","Skyblue","Gold"]

# Define Pyber color scheme
color_schema_df = pd.DataFrame(zip(city_types,colors),columns=["type","color"])
# Set "type" as index
color_schema_df.set_index("type",drop=True,inplace=True)

# Obtain the x and y coordinates for each of the three city types
# Group merged data by city type and city
pyber_city = pyber_df.groupby(["type","city"])

# Average Fare ($) Per City -> y_value
avr_fare_city = pyber_city["fare"].mean()

# Total Number of Rides Per City -> x_value
total_rides_city = pyber_city["ride_id"].count()

# Total Number of Drivers Per City -> bubble's size
total_drivers_city = pyber_city["driver_count"].max()

# Create subplot
fig,ax = plt.subplots() 
fig.suptitle("Piber Ride Sharing Data (2016)", y=0.95, fontsize=12, fontweight='bold')

# Build the scatter plots for each city types
for x in color_schema_df.index:
    x_axis = total_rides_city[x]
    y_axis = avr_fare_city[x]
    c=color_schema_df.loc[x]["color"] 
    size = [s*10 for s in total_drivers_city[x]]   
    plt.scatter(x_axis, y_axis, marker="o", facecolor=c, edgecolor="black",alpha=0.75, s=size, label=x, linewidths=1)

# Incorporate the other graph properties
x_max = total_rides_city.max()
y_max = avr_fare_city.max()
x_min = total_rides_city.min()
y_min = avr_fare_city.min()
plt.xlim(0, x_max+2)
plt.ylim(y_min-1.3, y_max+1)
# plt.title("Piber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid()

# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(title="City types", loc="best", fontsize="small", shadow=True, markerscale=0.6) 

# Incorporate a text label regarding circle size
plt_note = "Note: \nCircle size correlates with driver count per city"
ax.text(1.05, 0.65, plt_note,
        verticalalignment='center', horizontalalignment='left',
        transform=ax.transAxes,
        color='black', fontsize=10)

# Show plot
plt.show()

# Save scatter plot
fig.savefig("Images/PaberRideScatter.png", bbox_inches="tight")
```

## Total Fares by City Type

```python
# Calculate Type Percents

# Group merged data by "city type"
pyber_city_type = pyber_df.groupby(["type"])

# Calculate the Total Fares by City Type
total_fare_city_type = pyber_city_type["fare"].sum()

# Define Pyber fare Data Frame
total_fare = pd.DataFrame (columns=["type","color","total_fare"])
for x in total_fare_city_type.index:
    c=color_schema_df.loc[x]["color"]
    row=[x,c,total_fare_city_type[x]]
    total_fare.loc[len(total_fare)] = row

# set index    
total_fare.set_index("type",drop=True,inplace=True)

# Build Pie Chart
# Define size of figure
fig,ax = plt.subplots(figsize=(5.6,3.5))

# Define plot
explode = (0, 0, 0.1)
plt.pie (list(total_fare["total_fare"]), explode=explode,
         colors=list(total_fare["color"]),
         labels=list(total_fare.index),
         autopct="%1.1f%%",shadow=True,startangle=150)

# Define title
plt.title("% of Total Fares by City Type")

# Create axes 
plt.axis("tight")

# Show plot
plt.show

# Save Figure
plt.savefig("Images/PyberFaresPie.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Calculate the Total Rides by City Type
total_ride_city_type = pyber_city_type["ride_id"].count()

# Define Pyber fare Data Frame
total_ride = pd.DataFrame (columns=["type","color","total_rides"])
for x in total_ride_city_type.index:
    c=color_schema_df.loc[x]["color"]
    row=[x,c,total_ride_city_type[x]]
    total_ride.loc[len(total_ride)] = row

# set index    
total_ride.set_index("type",drop=True,inplace=True)

# Build Pie Chart
# Define size of figure
fig,ax = plt.subplots(figsize=(5.6,3.5))

# Define plot
explode = (0, 0, 0.1)

# Define plot
plt.pie (list(total_ride["total_rides"]), explode=explode,
         colors=list(total_ride["color"]),
         labels=list(total_ride.index),
         autopct="%1.1f%%",shadow=True,startangle=150)

# Define title
plt.title("% of Total Rides by City Type")

# Create axes 
plt.axis("tight")

# Show plot
plt.show

# Save Figure
plt.savefig("Images/PyberRidesPie.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

# Calculate total Drivers by city type
total_drivers_city_type = total_drivers_city.groupby(["type"]).sum()
#print(total_drivers_city_type.head())

# Define Pyber fare Data Frame
total_drivers = pd.DataFrame (columns=["type","color","total_drivers"])
for x in total_drivers_city_type.index:
    c=color_schema_df.loc[x]["color"]
    row=[x,c,total_drivers_city_type[x]]
    total_drivers.loc[len(total_drivers)] = row

# set index    
total_drivers.set_index("type",drop=True,inplace=True)

# Build Pie Chart
# Define size of figure
fig,ax = plt.subplots(figsize=(5.6,3.5))

# Define plot
explode = (0, 0, 0.1)
plt.pie (list(total_drivers["total_drivers"]), explode=explode,
         colors=list(total_drivers["color"]),
         labels=list(total_drivers.index),
         autopct="%1.1f%%",shadow=True,startangle=170)

# Define title
plt.title("% of Total Drivers by City Type")

# Create axes 
plt.axis("tight")

# total drivers
print("Total Drivers by City Type:")
for x in total_drivers_city_type.index:
   print(f"{x}: {total_drivers_city_type[x]}")

# Show plot
plt.show

# Save Figure
plt.savefig("Images/PyberDriversPie.png")
```

# Conclusions Based on Pyber Analysis

 * As a whole, there are much more drivers in the Urban area than in the Suburban and especially Rural areas, $80.9\%$ vs $16.5\%$ in the Suburban areas and $2.6\%$ in the Rural. They make the most rides $68.4\%$ and earn the most money $62.7\%$ comparing with Suburban and Rural areas.

 * However, perhaps due to a large number of the Drivers and Rides in the Urban areas and high level of competition, the Average Fare in this area is less than in the Suburban and especially in the Rural areas. Precisely $24.52\$$ vs $30.07\$$ the average fare in the Suburban and $34.62\$$ in the Rural areas. Therefore, the average ride in the Urban areas is cheaper than in the Suburban and Rural, but not drastic. The most expensive rides are in the Rural areas. It needs to mention that price range is significantly wide in the Rural areas, from a pretty cheap $24.12\$$ to the most expensive $43.28\$$, so the average fare in this area is not representative.

 * None the less, the total quantity of rides in the Urban area in $2.6$ times bigger than in the Suburban and in $12.9$ times bigger than in the Rural areas, that allows Urban drivers to earn $62.7\%$ of total fare, as Suburban drivers earn $30.5\%$ and Rural drivers earn $6.8\%$.

 * Despite the fact that the Urban areas out-performed the Suburban and Rural across all metrics, 1 Rural driver makes $0.07\%$ of the total rides and earns $0.09\%$ of the total fares. As well as 1 Suburban driver makes $0.05\%$ of the total rides and earns $0.06\%$ of the total fares and 1 Urban driver makes $0.03\%$ of the total rides and earns $0.03\%$ of the total fares. Either in a percentage $1\%$ of the Rural drivers make $2\%$ of the total rides and $2.6\%$ of the total fares. From the other hand $1\%$ Suburban drivers make accordingly $1.5\%$ and $1.8\%$, as well as $1\%$ of the Urban drivers make $0.8\%$ and $0.8\%$ accordingly. So that it seems like more profitable to work at the Rural areas, than in the Suburban and at least the least profitable to work in the Urban areas.
